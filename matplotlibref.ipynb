```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
# plt.show()
```

```python
plt.plot(x,y,'-r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) # rows, columns, plot number
plt.plot(x,y,'-r')

plt.subplot(1,2,2)
plt.plot(y,x,'bo')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
# axes: left, bottom, width, height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.15,0.4,0.3])
# Remember, axes: left, bottom, width, height

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

# Blue lines
for i in axes:
    i.plot(x,y)

# Orange lines
axes[0].plot(y,x*2)
axes[1].plot(y,x)
```

```python
fig = plt.figure(figsize=(3,2), dpi=50)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(3,2), dpi=200)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(3,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=10) # 0 is best location, default
```

```python
fig = plt.figure(figsize=(3,2), dpi=200)

ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color='green')
#ax.plot(x,y,color='#FF8C00') # RGB Hex
#ax.plot(x,y,color='purple',linewidth=3)
#ax.plot(x,y,color='blue',linewidth=5,alpha=0.2)
#ax.plot(x,y,color='purple',lw=3,linestyle='steps')
# Some types of ls: -- -. : steps -
ax.plot(x,y,color='red',lw=0.5,ls='steps',marker='o',markersize=4, markerfacecolor='yellow', markeredgewidth=2,markeredgecolor='purple')
# Some types of markers: o 1 *
```

```python
fig = plt.figure(figsize=(3,2), dpi=200)

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='red',lw=0.5,ls='steps',marker='o',markersize=4, markerfacecolor='yellow', markeredgewidth=2,markeredgecolor='purple')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
