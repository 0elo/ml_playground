# Principal Component Analysis

Principal component analysis (PCA) is an unsupervised statistical technique used to examine the interrelations among a set of variables in order to identify the underlying structure of those variables. Also closely related to factor analysis.

Where regression determines a line of best fit, factor analysis determines several orthogonal lines of best fit to the data set.

PCA is essentially just a transformation of your data and attempts to find out what features explain the most variance in your data.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
type(cancer)
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns = cancer['feature_names'])
```

```python
df.head()
```

### Using PCA to find what's important in seeing whether a tumor is malignant or benign

With 2 principal components

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
# PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c = cancer['target'], cmap = 'coolwarm')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns = cancer['feature_names'])
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```

With 3 principal components

```python
scaler2 = StandardScaler()
```

```python
scaler2.fit(df)
```

```python
scaled_data = scaler2.transform(df)
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(12,8))
ax = Axes3D(fig)
ax.scatter(x_pca[:,0], x_pca[:,1], x_pca[:,2], c = cancer['target'], cmap = 'coolwarm')
ax.set_xlabel('First Principal Component')
ax.set_ylabel('Second Principal Component')
ax.set_zlabel('Third Principal Component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns = cancer['feature_names'])
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df_comp,cmap='plasma')
```
